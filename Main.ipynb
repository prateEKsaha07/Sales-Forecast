```python
# imports
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("train.csv")
df = pd.DataFrame(data)
```

```python
new_data = pd.DataFrame(df[['Order Date','Segment','Region','Category','Sub-Category','Sales','Ship Date','Product Name']])
```

```python
new_data.head()
```

```python
new_data.head(50)
```

```python
# fro the Oder Date column 
new_data['Order Date'] = pd.to_datetime(new_data['Order Date'],format='%d/%m/%Y')
new_data['Day'] = new_data['Order Date'].dt.day
new_data['Month'] = new_data['Order Date'].dt.month
new_data['Year'] = new_data['Order Date'].dt.year
new_data['DayOFWeek'] = new_data['Order Date'].dt.dayofweek
new_data['Ship Date'] = pd.to_datetime(new_data['Ship Date'],format='%d/%m/%Y')
new_data["shippingDelays"] = (new_data['Ship Date'].dt.day - new_data['Order Date'].dt.day)
```

```python
new_data.info()
```

```python
new_data.head()
```

```python
new_data.drop(['Order Date'],inplace = True, axis = 1)
```

```python
new_data.head()
```

```python
new_data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
le  = LabelEncoder()

for col in ['Segment','Region','Category','Sub-Category','Product Name']:
    new_data[col] = le.fit_transform(new_data[col])

new_data.info()
```

```python
x = new_data[['Segment', 'Region', 'Category', 'Sub-Category','Day','Month','Year','DayOFWeek','shippingDelays','Product Name']]
y = new_data['Sales']
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test,y_train ,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
x_train.shape, x_test.shape,y_train.shape,y_test.shape

# SubCategory avg
avg_map = y_train.groupby(x_train['Sub-Category']).mean()

x_train['SubCategory_avg_sales'] = x_train['Sub-Category'].map(avg_map)
x_test['SubCategory_avg_sales'] = x_test['Sub-Category'].map(avg_map)

x_test['SubCategory_avg_sales'] = x_test['SubCategory_avg_sales'].fillna(avg_map.mean())

# Product avg
avg_product_map = y_train.groupby(x_train['Product Name']).mean()

x_train['product_avg_sales'] = x_train['Product Name'].map(avg_product_map)
x_test['product_avg_sales'] = x_test['Product Name'].map(avg_product_map)

x_test['product_avg_sales'] = x_test['product_avg_sales'].fillna(avg_product_map.mean())

x_train = x_train.drop('Product Name', axis=1)
x_test = x_test.drop('Product Name', axis=1)

x_test.info()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

lr = LinearRegression()
lr.fit(x_train, y_train)

preds = lr.predict(x_test)

print("MAE:", mean_absolute_error(y_test, preds))
print("MSE:", mean_squared_error(y_test, preds))
print("R2 Score:", r2_score(y_test, preds))
```

```python
from sklearn.ensemble import RandomForestRegressor

rf = RandomForestRegressor()
rf.fit(x_train,y_train)

rf_preds = rf.predict(x_test)

print("MAE:", mean_absolute_error(y_test, rf_preds))
print("MSE:", mean_squared_error(y_test, rf_preds))
print("R2 Score:", r2_score(y_test, rf_preds))
```

```python
from xgboost import XGBRegressor

model = XGBRegressor(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=6,
    random_state=42
)

model.fit(x_train, y_train)
xgb_preds = model.predict(x_test)
print("MAE:", mean_absolute_error(y_test, xgb_preds))
print("MSE:", mean_squared_error(y_test, xgb_preds))
print("R2 Score:", r2_score(y_test, xgb_preds))
```
